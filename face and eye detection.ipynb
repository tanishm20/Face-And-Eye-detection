# Face And Eye Detection by Image

```python
import cv2,time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
img = cv2.imread("harrypotter.jpg",1)
face= cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye= cv2.CascadeClassifier("haarcascade_eye.xml")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
faces = face.detectMultiScale(img,scaleFactor=1.05,minNeighbors=5)
eyes = eye.detectMultiScale(img,scaleFactor=1.05,minNeighbors=5)

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(225,255,0),2)
    gray = img[y:y+h, x:x+w]
    color = img[y:y+h, x:x+w]
eyes = eye.detectMultiScale(gray)
for (x,y,w,h) in eyes:
    cv2.rectangle(color,(x,y),(x+w,y+h),(25,225,255),2)
plt.imshow(img)
plt.show()
```

# Face Detection By Web CAM

```python
video = cv2.VideoCapture(0)
face= cv2.CascadeClassifier("haarcascade_frontalface_default.xml")


while(True):
    ret,frame=video.read()
    clr=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    clr=cv2.GaussianBlur(clr,(21,21),0)
    faces = face.detectMultiScale(clr,scaleFactor=1.05,minNeighbors=5)
    
    for x,y,w,h in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow('capturing',frame)
    if cv2.waitKey(1)==27:
        break
        
        

video.release()
cv2.destroyAllWindows()
```
